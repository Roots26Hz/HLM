# Part I — NumPy for Machine Learning

## 1.1 Creating Arrays

```python
# 1. 4x4 identity and a 3x5 array of all 7s (single call each)
ident = np.eye(4)
sevens = np.full((3, 5), 7)
print(ident)
print(sevens)

# 2. 10 numbers evenly spaced between -1 and 1
print(np.linspace(-1, 1, 10))

# 3. Convert a Python list to an array and print its dtype
arr = np.array([2, 4, 6, 8])
print(arr, arr.dtype)
```

## 1.2 Array Attributes & Reshaping

```python
# 1. 0..23 -> 4x6 -> 2x3x4
a = np.arange(24)
m46 = a.reshape(4, 6)
m234 = a.reshape(2, 3, 4)
print(m46.shape, m234.shape)

# 2. Transpose the 4x6, printing shape before/after
print('before:', m46.shape)
mt = m46.T
print('after :', mt.shape)

# 3. Flatten back to 1-D and confirm length 24
flat = mt.ravel()
print('length:', flat.size)
```

## 1.3 Indexing, Slicing & Boolean Masking

```python
X = np.arange(25).reshape(5, 5)

# 1. middle 3x3 sub-matrix
print(X[1:4, 1:4])

# 2. all even numbers via boolean mask
print(X[X % 2 == 0])

# 3. replace every element > 15 with -1
Xc = X.copy()
Xc[Xc > 15] = -1
print(Xc)
```

## 1.4 Arithmetic Operations & Broadcasting

```python
# 1. flat discount of 10 via broadcasting (no loop)
prices = np.array([100, 250, 80, 500])
print(prices - 10)

# 2. 4x3 ones + row vector [1,2,3]; predicted shape (4, 3)
M = np.ones((4, 3))
res = M + np.array([1, 2, 3])
print('predicted shape (4, 3); actual:', res.shape)
print(res)

# 3. standardize an array (subtract mean, divide by std)
v = np.array([2, 4, 4, 4, 5, 5, 7, 9], dtype=float)
print((v - v.mean()) / v.std())
```

## 1.5 Aggregations & Axis-wise Statistics

```python
R = np.random.randint(0, 100, (5, 4))
print('matrix:\n', R)

# 1. mean of every column, max of every row
print('col means:', R.mean(axis=0))
print('row maxes:', R.max(axis=1))

# 2. index of overall minimum
print('argmin (flat):', R.argmin())
print('argmin (row, col):', np.unravel_index(R.argmin(), R.shape))

# 3. range (max - min) per column
print('col ranges:', R.max(axis=0) - R.min(axis=0))
```

## 1.6 Stacking, Concatenation & Splitting

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# 1. vstack -> 4x2, hstack -> 2x4
print(np.vstack([A, B]))
print(np.hstack([A, B]))

# 2. split 12 elements into 3 equal parts
for part in np.split(np.arange(12), 3):
    print(part)

# 3. column_stack three length-5 feature arrays -> 5x3
age = np.array([21, 22, 23, 24, 25])
height = np.array([160, 165, 170, 175, 180])
weight = np.array([55, 60, 65, 70, 75])
print(np.column_stack([age, height, weight]))
```

## 1.7 Linear Algebra Essentials

```python
# 1. predictions X @ w
X = np.array([[1, 1], [1, 2], [1, 3], [1, 4]])
w = np.array([2, 3])
print('X @ w =', X @ w)

# 2. L1 and L2 norms
u = np.array([3, -4, 0, 12])
print('L1:', np.linalg.norm(u, ord=1), '  L2:', np.linalg.norm(u, ord=2))

# 3. verify X X^-1 = I for a 3x3 invertible matrix
P = np.array([[2., 1., 1.], [1., 3., 2.], [1., 0., 0.]])
print('X X^-1 == I :', np.allclose(P @ np.linalg.inv(P), np.eye(3)))
```

## 1.8 Random Module for ML

```python
# 1. seed 7, 3x3 randint 1..100, run twice -> identical
np.random.seed(7); first = np.random.randint(1, 101, (3, 3))
np.random.seed(7); second = np.random.randint(1, 101, (3, 3))
print('identical:', np.array_equal(first, second))

# 2. 4x2 weight matrix from standard normal
print(np.random.randn(4, 2))

# 3. shuffle arange(20), 80/20 split
idx = np.arange(20)
np.random.shuffle(idx)
train_idx, test_idx = idx[:16], idx[16:]
print('train:', train_idx)
print('test :', test_idx)
```

## 1.9 Sorting, Searching & Unique Values

```python
# 1. indices that sort scores in descending order
scores = np.array([0.2, 0.91, 0.4, 0.88, 0.05])
print(np.argsort(scores)[::-1])

# 2. unique labels and their counts
labels = np.array([0, 1, 2, 1, 0, 2, 2, 1, 0])
values, counts = np.unique(labels, return_counts=True)
print(dict(zip(values.tolist(), counts.tolist())))

# 3. threshold probabilities at 0.5 -> binary predictions
p = np.array([0.3, 0.7, 0.45, 0.9, 0.1])
print(np.where(p >= 0.5, 1, 0))
```

## 2.2 Loading & Exploring a Real Dataset

```python
# 1. shape and first 5 rows
print(wine_df.shape)
print(wine_df.head())

# 2. info(): column count and missing values
wine_df.info()
print('columns:', wine_df.shape[1], '| total missing:', int(wine_df.isnull().sum().sum()))

# 3. feature with the largest range
desc = wine_df.describe()
ranges = desc.loc['max'] - desc.loc['min']
print('largest range:', ranges.drop('target').idxmax())
```

## 2.3 Selecting & Filtering Data

```python
# 1. alcohol & color_intensity for first 10 rows
print(wine_df[['alcohol', 'color_intensity']].head(10))

# 2. loc rows 5..10 inclusive, magnesium only
print(wine_df.loc[5:10, 'magnesium'])

# 3. alcohol > 14 AND target == 0
subset = wine_df[(wine_df['alcohol'] > 14) & (wine_df['target'] == 0)]
print(subset.shape)
```

## 2.4 Handling Missing Data

```python
# 1. set 5 random magnesium values to NaN
dfm = wine_df.copy()
rng = np.random.default_rng(0)
nan_idx = rng.choice(dfm.index, size=5, replace=False)
dfm.loc[nan_idx, 'magnesium'] = np.nan

# 2. count missing per column
print(dfm.isnull().sum()[dfm.isnull().sum() > 0])

# 3. fill with median; confirm zero missing
dfm['magnesium'] = dfm['magnesium'].fillna(dfm['magnesium'].median())
print('remaining missing:', int(dfm['magnesium'].isnull().sum()))
```

## 2.5 Feature Engineering: apply & map

```python
dfe = wine_df.copy()

# 1. phenol_ratio = flavanoids / total_phenols
dfe['phenol_ratio'] = dfe['flavanoids'] / dfe['total_phenols']

# 2. high_alcohol via apply
dfe['high_alcohol'] = dfe['alcohol'].apply(lambda x: 1 if x > 13 else 0)

# 3. map numeric target -> class names
from sklearn.datasets import load_wine
names = load_wine().target_names
dfe['class_name'] = dfe['target'].map({i: n for i, n in enumerate(names)})
print(dfe[['phenol_ratio', 'high_alcohol', 'class_name']].head())
```

## 2.6 Sorting Values

```python
# 1. top 5 by alcohol descending
print(wine_df.sort_values('alcohol', ascending=False).head(5)[['alcohol', 'target']])

# 2. sort by target asc, then color_intensity desc
print(wine_df.sort_values(['target', 'color_intensity'], ascending=[True, False]).head())

# 3. 3 samples with the lowest hue
print(wine_df.nsmallest(3, 'hue')[['hue']])
```

## 2.7 GroupBy & Aggregation

```python
# 1. mean of every feature per class
print(wine_df.groupby('target').mean())

# 2. alcohol mean/min/max per class in one agg
print(wine_df.groupby('target')['alcohol'].agg(['mean', 'min', 'max']))

# 3. class with highest average color_intensity (by code)
print('highest avg color_intensity -> class',
      wine_df.groupby('target')['color_intensity'].mean().idxmax())
```

## 2.8 Merging & Concatenating DataFrames

```python
# 1. split first 5 rows into two column-subsets, recombine column-wise
head5 = wine_df.head(5)
left = head5[['alcohol', 'malic_acid']]
right = head5[['hue', 'proline']]
print(pd.concat([left, right], axis=1))

# 2. two 3-row DataFrames concatenated row-wise
d1 = pd.DataFrame({'x': [1, 2, 3]})
d2 = pd.DataFrame({'x': [4, 5, 6]})
print(pd.concat([d1, d2], ignore_index=True))

# 3. lookup target -> quality_tier, merged onto first 10 rows
lookup = pd.DataFrame({'target': [0, 1, 2], 'quality_tier': ['A', 'B', 'C']})
print(pd.merge(wine_df.head(10), lookup, on='target', how='left')[['target', 'quality_tier']])
```

## 2.9 Encoding Categorical Variables

```python
# 1. one-hot encode a grade column
grades = pd.DataFrame({'grade': ['A', 'B', 'A', 'C', 'B']})
print(pd.get_dummies(grades, columns=['grade']))

# 2. label-encode (grades have a natural order)
print(grades['grade'].map({'A': 2, 'B': 1, 'C': 0}))

# 3. one-hot encode the Wine target.
# Correct because the three classes have NO ordinal relationship, so label
# encoding would imply a false ordering (class 2 > class 1 > class 0).
print(pd.get_dummies(wine_df['target'], prefix='class').head())
```

## 2.10 Correlation Analysis & Train/Test Sampling

```python
# 1. most strongly correlated pair of features
features = wine_df.drop(columns='target')
corr = features.corr().abs()
corr_vals = corr.values.copy()  # .values can be read-only in recent pandas
np.fill_diagonal(corr_vals, 0)
i, j = np.unravel_index(corr_vals.argmax(), corr_vals.shape)
print('most correlated pair:', corr.index[i], '&', corr.columns[j])

# 2. 70/30 split via sample + iloc
shuffled = wine_df.sample(frac=1, random_state=0).reset_index(drop=True)
cut = int(len(shuffled) * 0.7)
train, test = shuffled.iloc[:cut], shuffled.iloc[cut:]
print('train:', train.shape, '| test:', test.shape)

# 3. feature most correlated with the target
tcorr = wine_df.corr()['target'].drop('target').abs()
print('most predictive feature:', tcorr.idxmax())
```

# Part III — Capstone: End-to-End Preprocessing of the Wine Dataset

Using only NumPy and Pandas, mirroring a real preprocessing pipeline.

```python
# 1. Load + inspect
df = load_wine_df()
print('shape:', df.shape)
print(df.dtypes.value_counts())
df.describe()
```

```python
# 2. Introduce 10 random missing values across feature columns
cap = df.copy()
feature_cols = [c for c in cap.columns if c != 'target']
rng = np.random.default_rng(42)
for _ in range(10):
    r = rng.integers(0, len(cap))
    c = feature_cols[rng.integers(0, len(feature_cols))]
    cap.iloc[r, cap.columns.get_loc(c)] = np.nan
print('missing per column (nonzero):')
print(cap.isnull().sum()[cap.isnull().sum() > 0])
```

```python
# 3. Impute every missing value with its column median
cap[feature_cols] = cap[feature_cols].fillna(cap[feature_cols].median())
print('total missing after impute:', int(cap.isnull().sum().sum()))
```

```python
# 4. Engineer one new feature
cap['phenol_ratio'] = cap['flavanoids'] / cap['total_phenols']
feature_cols = feature_cols + ['phenol_ratio']
print(cap['phenol_ratio'].describe())
```

```python
# 5. Standardize feature columns with NumPy (reused helper does mean/std)
Xstd = standardize(cap[feature_cols].values)
print('standardized shape:', Xstd.shape)
```

```python
# 6. One-hot encode the target into three binary columns
target_ohe = pd.get_dummies(cap['target'], prefix='class').astype(int)
print(target_ohe.head())
```

```python
# 7. Shuffle (fixed seed) and split 80/20 using the shared helper
std_df = pd.DataFrame(Xstd, columns=feature_cols)
std_df = pd.concat([std_df, target_ohe], axis=1)
train_df, test_df = train_test_split_df(std_df, test_size=0.2, seed=42)
X_train = train_df[feature_cols].values
X_test = test_df[feature_cols].values
print('X_train:', X_train.shape, '| X_test:', X_test.shape)
```

```python
# 8. Confirm standardized features have mean ~0 and std ~1
print('overall mean ~', round(float(Xstd.mean()), 6))
print('overall std  ~', round(float(Xstd.std()), 6))
```
